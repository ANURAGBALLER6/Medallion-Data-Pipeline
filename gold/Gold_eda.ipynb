```python
import matplotlib.pyplot as plt
import seaborn as sns

top5 = df_city_kpis.nlargest(5, "total_pickups")

plt.figure(figsize=(8,5))
sns.barplot(x="city", y="total_pickups", data=top5, palette="viridis")
plt.title("Top 5 Cities by Total Pickups", fontsize=14, fontweight="bold")
plt.xlabel("City", fontsize=12)
plt.ylabel("Total Pickups", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The above chart shows that top 2 cities have almost same number of pickups and there is a slight drop in the rest 3
cities.

```python
plt.figure(figsize=(8,5))
sns.histplot(df_daily_kpis["avg_revenue_per_trip_usd"], bins=20, kde=True, color="steelblue")
plt.title("Distribution of Avg Revenue per Trip (USD)", fontsize=14, fontweight="bold")
plt.xlabel("Avg Revenue per Trip (USD)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

The histogram shows the distribution of the average revenue per trip in USD. From the plot, we can observe that the distribution appears to be roughly normally distributed (or bell-shaped).

The peak of the histogram is around 45-55 USD, indicating that the most frequent average revenue per trip falls within this range.
The frequencies decrease as the average revenue per trip moves away from the peak in both directions.
The distribution is relatively symmetrical, although there might be a slight skew.
There are a few data points with very low or very high average revenue per trip, but the majority of the data is concentrated around the mean.
Overall, this suggests that most trips generate an average revenue within a predictable range, with fewer trips generating significantly higher or lower average revenue.

```python
corr = df_driver_stats.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Driver Stats Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

Looking at the correlation heatmap we generated for the df_driver_stats DataFrame, here are the notable correlations:

There is a strong positive correlation (0.79) between total_trips and total_earnings_usd. This makes sense, as drivers who complete more trips are likely to earn more.
There is a strong negative correlation (-0.53) between avg_trip_fare_usd and avg_tip_rate. This suggests that as the average fare per trip increases, the average tip rate tends to decrease.
There is a moderately positive correlation (0.64) between avg_tip_usd and avg_tip_rate. This is also expected, as a higher tip rate generally leads to a higher average tip amount.
The global_avg_tip_usd column shows no correlation with any other variable, which is likely because it's a constant value across all drivers in this dataset.
